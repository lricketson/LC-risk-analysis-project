```python
import pandas as pd
from main import specialty_preprocess_df

my_df = pd.read_csv(
    "./datasets/lc_data_2007_to_2018.csv",
    low_memory=False,
    encoding="latin1",
    nrows=100000,  # only looking at 100k rows right now for performance
)
pd.set_option("display.max_columns", None)
cleaned_df = specialty_preprocess_df(my_df)

cleaned_df["net_pnl"] = cleaned_df["total_pymnt"] - cleaned_df["loan_amnt"]
cleaned_df["loan_to_income"] = cleaned_df["loan_amnt"] / cleaned_df["annual_inc"]
cleaned_df["payment_to_income"] = cleaned_df["installment"] / (
    cleaned_df["annual_inc"] / 12
)


cleaned_df.to_csv("cleaned_df.csv")
```
