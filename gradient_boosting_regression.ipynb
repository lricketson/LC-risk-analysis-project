```python
import pandas as pd

cleaned_df = pd.read_csv("cleaned_df.csv", index_col=0)
```

```python
from main import prepare_ml_data

X_train, X_test, y_train, y_test = prepare_ml_data(cleaned_df, scale=False)
```

```python
from sklearn.ensemble import HistGradientBoostingClassifier
from sklearn.calibration import CalibratedClassifierCV

gb_model = HistGradientBoostingClassifier(random_state=42)
gb_model.fit(X=X_train, y=y_train)

calibrated_gb = CalibratedClassifierCV(gb_model, method="isotonic", cv=5)
calibrated_gb.fit(X_train, y_train)

gb_predicted_probs = calibrated_gb.predict_proba(X_test)[:, 1]

export_df = pd.DataFrame({"gb_prob": gb_predicted_probs}, index=X_test.index)
export_df.to_csv("gb_predictions.csv")
print("Gradient Boosting predictions successfully exported!")
```
