```python
import numpy as np
import pandas as pd

PERCENTAGE_CONSTANT = 100
NAN_CUTOFF = 0.95

df = pd.read_csv(
    "./datasets/lc_data_2007_to_2018.csv",
    low_memory=False,
    encoding="latin1",
    nrows=10000,  # only looking at 10k rows right now for performance
)
pd.set_option("display.max_columns", None)
dropped_cols_list = ["emp_title", "url", "zip_code"]
for col in df.columns:
    number_of_nans = df[col].isna().sum()
    col_length = len(df)
    ratio_of_nans = number_of_nans / col_length
    if ratio_of_nans >= NAN_CUTOFF:
        dropped_cols_list.append(col)
df = df.drop(columns=dropped_cols_list)
```
